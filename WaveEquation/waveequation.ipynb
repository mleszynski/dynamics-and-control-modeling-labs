# Problem 1
Numerically approximate the solution to the following BVP

$$
u_{tt} = u_{xx} \\
u(0,t) = u(1,t) = 0, \\
u(x,0) = \sin(2\pi x), \\
u_t(x,0) = 0. 
$$

Numerically approximate the solution $u(x,t)$ for $t\in[0,0.5]$. Use $J=50$ subintervals in the $x$ dimension and $M=50$ subintervals in the $t$ dimension. Animate the results. Compare you results with the analytic solution $u(x,t) = \sin(2\pi x)\cos(2\pi t).$ This function is know as a standing wave.

```python
plt.ioff()

# define figure params ###############################################################
delta_x = 1./50
delta_t = 0.5/50
lamb = delta_t/delta_x
J = np.linspace(0, 1.0, 51)
M = np.arange(0, 0.5, delta_t)
fig = plt.figure()
fig.suptitle('Numerical Solution')
ax = fig.add_subplot(111)
ax.set_xlim((0, 1))
ax.set_ylim((-1, 1))
line, = plt.plot([], [], color='r')

# calculate coordinates #############################################################
f_x = lambda x:np.sin(2*np.pi*x)

    # create U_0, U_1 ###############################################################
U = np.array([f_x(j) for j in J])
beginning = U[0]
end = U[-1]
U_next = np.array([U[i] + (lamb**2)/2*(U[i-1]-2*U[i]+U[i+1]) for i in range(1,len(J)-1)])
U_next = np.hstack((U[0], U_next))
U_next = np.hstack((U_next, U[-1]))
U = np.vstack((U, U_next))

    # create rest of U_j #############################################################
A = np.diag(np.full(49, 2*(1-lamb**2)))+np.diag(np.full(48, lamb**2), 1)+np.diag(np.full(48, lamb**2), -1)

for i in range(2, 50):
    U_next = A@U[i-1][1:-1]-U[i-2][1:-1]
    U_next = np.hstack((beginning, U_next))
    U_next = np.hstack((U_next, end))
    U = np.vstack((U, U_next))

# embed animation ###################################################################
def update(i):
    line.set_data(J, U[i])
    return U

ani = animation.FuncAnimation(fig, update, frames=range(50), interval=25)
HTML(ani.to_html5_video())
```

```python
plt.ioff()

# define figure params ###############################################################
J = np.arange(0, 1.0, 1./50)
M = np.arange(0, 0.5, 0.5/50)
fig = plt.figure()
fig.suptitle('Analytic Solution')
ax = fig.add_subplot(111)
ax.set_xlim((0, 1))
ax.set_ylim((-1, 1))
line, = plt.plot([],[], color='b')

# calculate coordinates #############################################################
def analytic(x,t):
    return np.sin(2*np.pi*x)*np.cos(2*np.pi*t)

anim = []
for m in range(len(M)):
    temp = []
    for j in range(len(J)):
        temp.append(analytic(J[j],M[m]))
    anim.append(temp)

# embed animation ###################################################################
def update(i):
    line.set_data(J, anim[i])
    return anim

ani = animation.FuncAnimation(fig, update, frames=range(50), interval=25)
HTML(ani.to_html5_video())
```

# Problem 2

Consider the boundary value problem 

$$
u_{tt} = u_{xx}, \\
u(0,t) = u(1,t) = 0, \\
u(x,0) = 0.2e^{-m^2(x-1/2)^2} \\
u_t(x,0) = 0.4m^2(x-\frac{1}{2})e^{-m^2(x-1/2)^2}.
$$

The solution of this problem is ta Gaussian pulse. It travels to the right at a constant speed. This solution models, for example, a wave pulse in a stretched string. Note that the fixed boundary conditions reflect the pulse back when it meets the boundary.

Numerically approximate the solution $u(x,t)$ for $t\in[0,1]$. Set $m=20$. Use 200 subintervals in space and 220 in time, and animate your results. Then use 200 subintervals in space and 180 in time, and animate your results. Note that the stability condition is not satisfied for the second mesh.

# Problem 3

Consider the initial boundary value problem
$$
	u_{tt} = u_{xx}, \\
	u(0,t) = u(1,t) = 0, \\
	u(x,0) = 0.2e^{-m^2(x-1/2)^2}\\
	u_t(x,0) = 0.
$$
The initial condition separates into two smaller, slower-moving pulses, one travelling to the right and the other to the left.
This solution models, for example, a plucked guitar string

Numerically approximate the solution $u(x,t)$ for $t \in \left[0,2\right]$.
Set $m=20$.
Use 200 subintervals in space and 440 in time, and animate your results.
It is rather easy to see that the solution to this problem is the sum of two travelling waves, one travelling to the left and the other to the right, as described earlier.

# Problem 4

Consider the initial boundary value problem

$$
	u_{tt} = u_{xx}, \\
	u(0,t) = u(1,t) = 0, \\
	u(x,0) = \begin{cases} 1/3 & \text{if } 5/11 < x < 6/11,\\
	0 & \text{otherwise}
	\end{cases}\\
	u_t(x,0) = 0.
$$

Numerically approximate the solution $u(x,t)$ for $t \in \left[0, 2\right]$.
Use 200 subintervals in space and 440 in time, and animate your results.
Even though the method is second order and stable for this discretization, since the initial condition is discontinuous there are large dispersive errors.

```python
plt.ioff()

# define figure params ###############################################################
delta_x = 1.0/200
delta_t = 2.0/440
lamb = delta_t/delta_x
J = np.arange(0, 1.0, delta_x)
M = np.arange(0, 2.0, delta_t)
fig = plt.figure()
fig.suptitle('200 space-intervals, 440 time-intervals')
ax = fig.add_subplot(111)
ax.set_xlim((0, 1))
ax.set_ylim((-1, 1))
line, = plt.plot([], [], color='r')

# calculate coordinates #############################################################
def f_x(x):
    if (x>(5.0/11) and x<(6.0/11)):
        return 1.0/3
    else:
        return 0.0

    # create U_0, U_1 ###############################################################
U = np.array([f_x(j) for j in J])
U_next = np.array([U[i]+(lamb**2)/2*(U[i-1]-2*U[i]+U[i+1]) for i in range(1,len(J)-1)])
U_next = np.hstack((np.array([0]), U_next))
U_next = np.hstack((U_next, np.array([0])))
U = np.vstack((U, U_next))

    # create rest of U_j #############################################################
A = np.diag(np.full(200, 2*(1-lamb**2)))+np.diag(np.full(199, lamb**2), 1)+np.diag(np.full(199, lamb**2), -1)
for i in range(2, 440):
    U_next = A@U[i-1]-U[i-2]
    U = np.vstack((U, U_next))

# embed animation ###################################################################
def update(i):
    line.set_data(J, U[i])
    return U

ani = animation.FuncAnimation(fig, update, frames=range(440), interval=25)
HTML(ani.to_html5_video())
```

# Problem 5


Numerically solve the initial value problem

$$
	u_t -su_x + uu_x = u_{xx}, \quad x \in (-\infty,\infty),\\
	u(x,0) = v(x),
$$

for $t \in [0,1]$.
Let the perturbation $v(x)$ be given by

$$
v(x) = 3.5(\sin{(3x)} + 1)\frac{1}{\sqrt{2\pi}} \exp{(-x^2/2)}
$$

And let the initial condition be $u(x, 0) = \hat{u}(x) + v(x)$
Approximate the $x$ domain,$(-\infty, \infty)$, numerically by the finite interval $[-20,20]$, and fix $u(-20) = u_-$, $u(20) = u_+$. Let $u_- = 5$, $u_+ = 1$.
Use 150 intervals in space and 350 steps in time.
Animate your results.
You should see the solution converge to a translate of the travelling wave $\hat{u}$.

Hint: This difference scheme is no longer a linear equation.
We have a nonlinear equation in $U^{n+1}$.
We can still solve this function using Newton's method or some other similar solver.
In this case, use `scipy.optimize.fsolve`.

```python
plt.ioff()

# define figure params ###############################################################
delta_x = 40./150
delta_t = 1.0/350
u_neg = 5.
u_pos = 1.
s = (u_neg + u_pos)/2
a = (u_neg - u_pos)/2
K_1 = delta_t/(4*delta_x)
K_2 = delta_t/(2*(delta_x**2))
J = np.arange(-20, 20, delta_x)
fig = plt.figure()
fig.suptitle('150 space-intervals, 350 time-intervals')
ax = fig.add_subplot(111)
ax.set_xlim((-20, 20))
ax.set_ylim((0, 6))
line, = plt.plot([], [], color='r')

# calculate coordinates #############################################################
    # create U_0 ####################################################################
v_x = lambda x:3.5*(np.sin(3.0*x)+1)*(1/np.sqrt(2*np.pi))*np.exp(-(x**2)/2)
f_x = lambda x:s-a*np.tanh((a*x)/2)+v_x(x)
U = np.array([f_x(j) for j in J])
U[0] = u_neg
U[-1] = u_pos
U = np.vstack((U, np.copy(U)))

def get_next_u(U_n1, U_n, s, K_1, K_2):
    sol = K_1*((s-U_n1[1:-1])*(U_n1[2:]-U_n1[0:-2])+(s-U_n[1:-1])*(U_n[2:]-U_n[0:-2]))\
        + K_2*((U_n1[2:]-2*U_n1[1:-1]+U_n1[0:-2])+(U_n[2:]-2*U_n[1:-1]+U_n[0:-2]))\
        + U_n[1:-1]
    sol = np.hstack((np.array([5.]),sol))
    sol = np.hstack((sol, np.array([1.])))
    return sol-U_n1

    # create rest of U_j ############################################################
for i in range(1, 350):
    guess = np.copy(U[-1])
    root = opt.fsolve(get_next_u, guess ,args=(guess, s, K_1, K_2))
    U = np.vstack((U, root))

# embed animation ###################################################################
def update(i):
    line.set_data(J, U[i])
    return U

ani = animation.FuncAnimation(fig, update, frames=range(350), interval=25)
HTML(ani.to_html5_video())
```
